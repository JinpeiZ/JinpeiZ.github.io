## Build a "dashboard" interface for the buildings dataset

### Left Component

```python
import pandas as pd
import numpy as np
import bqplot
import traitlets
import ipywidgets
import matplotlib.pyplot as plt
%matplotlib inline

buildings = pd.read_csv('building_inventory.csv') # read dataset
buildings.dropna()
buildings # take a quick look at the dataset
```

```python
# build left component
# 1. Data
# calculate the sum of total square footage for different criteria
totalSum = buildings.groupby(['Congress Dist', 'Agency Name'])['Square Footage'].sum().reset_index(name='Sum of total square footage')
df = pd.DataFrame(totalSum)
df.sort_values(by='Congress Dist')
df
```

```python
row_unique = df['Congress Dist'].unique()
col_unique = df['Agency Name'].unique()
data = pd.DataFrame(index = row_unique, columns = col_unique)
for i in df.index:
    data.loc[df['Congress Dist'][i],df['Agency Name'][i]] = df['Sum of total square footage'][i]
data = data.fillna(0)
data
```

```python
# 2. Scales
x_sc = bqplot.OrdinalScale()
y_sc = bqplot.OrdinalScale()
col_sc = bqplot.ColorScale(scheme="Blues")

# 3. Axis
ax_x = bqplot.Axis(scale = x_sc, label='Governmental department')
ax_y = bqplot.Axis(scale = y_sc, orientation='vertical', label='Congressional District')
ax_col = bqplot.ColorAxis(scale = col_sc, orientation='vertical', side='right')

# 4. Mark
heat_map = bqplot.GridHeatMap(color=np.array(data),
                            scales={'color':col_sc, 'row':y_sc, 'column':x_sc},
                            interactions={'click':'select'},
                            anchor_style={'fill':'pink'},
                            selected_style={'opacity':1.0},
                            unselected_style={'opacity':0.8})

# build the gridheatmap
fig = bqplot.Figure(marks = [heat_map], axes=[ax_col, ax_x, ax_y])
```

### Right Component

```python
# 1. Data
groupYear = buildings.groupby(['Year Acquired'])['Square Footage'].sum().reset_index(name='Total Footage')

# 2. Scales
x_scl = bqplot.OrdinalScale()
y_scl = bqplot.LinearScale()

# 3. Axis
x_axl = bqplot.Axis(scale=x_scl, label='Year')
y_axl = bqplot.Axis(scale=y_scl, label='Total Square Footage Acquired that Year', orientation='vertical')

# 4. Marks
# set region mask
region_mask = (groupYear['Year Acquired']>=1900) & (groupYear['Year Acquired']<=2000)\
& (groupYear['Total Footage']>=10000) & (groupYear['Total Footage']<=400000)

linePlot = bqplot.Lines(x = groupYear['Year Acquired'][region_mask],\
                        y = groupYear['Total Footage'][region_mask],\
                        scales={'x':x_scl, 'y':y_scl}, line_style='solid')

# build the lineplot
figl = bqplot.Figure(marks = [linePlot], axes = [x_axl, y_axl])
```

### Link left and right components and display the dashboard

```python
# 5. Interactions and Linking
mySelectedLabel = ipywidgets.Label() # a widget label
def on_selected(change):
    if len(change['owner'].selected) == 1:
        i, j = change['owner'].selected[0]
        print(i,j)
        v = data.iloc[i][j]
        mySelectedLabel.value = 'Sum of Total Square Footage = ' + str(v)
        # link to the line plot
        dfl = buildings[buildings['Congress Dist']==i]
        dfline = buildings[buildings['Agency Name']==col_unique[j]]
        groupYear = dfline.groupby(['Year Acquired'])['Square Footage'].sum().reset_index(name='Total Footage')
        # set region mask
        region_mask = (groupYear['Year Acquired']>=1900) & (groupYear['Year Acquired']<=2000)\
        & (groupYear['Total Footage']>=10000) & (groupYear['Total Footage']<=400000)
        print(groupYear.loc[region_mask])
        linePlot.x = groupYear['Year Acquired'][region_mask]
        linePlot.y = groupYear['Total Footage'][region_mask]

# link heatmap with the function
heat_map.observe(on_selected, 'selected')

# change line plot style
lineStyle = ['solid', 'dashed', 'dotted', 'dash_dotted']
lineStylePicker = ipywidgets.Dropdown(options=lineStyle, value=lineStyle[0], disabled=False)
ipywidgets.link((lineStylePicker,'value'),(linePlot,'line_style'))

# create the linked dashboard
fig.layout.min_width='700px'
figl.layout.min_width='700px'
figures = ipywidgets.HBox([fig,figl])
picker = ipywidgets.VBox([mySelectedLabel,lineStylePicker])
myDashboard = ipywidgets.VBox([picker, figures])
myDashboard
```
